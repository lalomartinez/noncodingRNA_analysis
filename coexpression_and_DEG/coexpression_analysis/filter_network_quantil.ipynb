```python
import pandas as pd
```

```python
# Load the BIG network file
big_network_file = "/media/eduardo/D1/ncRNAs_Leishmania_spp/RNAseq/Lmajor/analysis/06_WGCNA/tables/bigNet02_edges.txt"
big_network_df = pd.read_csv(big_network_file, sep="\t")
```

```python
quant = big_network_df['weight'].quantile([0.95]).tolist()
```

```python
filtered_network_df_quantil_filtered = big_network_df[big_network_df['weight'] >= quant[0]]
```

```python
# Save the filtered network to a new file
filtered_network_df_quantil_filtered.to_csv("/media/eduardo/D1/ncRNAs_Leishmania_spp/RNAseq/Lmajor/analysis/06_WGCNA/tables/bigNet02_edges_filteredby085quantil.txt", sep="\t", index=False)
```
